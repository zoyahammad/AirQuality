# Hourly Analysis

Converted the data time into hours, in order to see hourly trend of PM 2.5's in the three different cities.

```python
# hourly analysis 
df1['Date'] = pd.to_datetime(df1['Date'])
df1['Hour'] = df1['Date'].dt.hour
hourly_pm25de = df1.groupby('Hour')['PM2.5, ug/m3'].mean()

hourly_pm25de.plot(kind='line')
plt.ylabel('PM2.5, ug/m3')
plt.title ("Hourly PM 2.5's Level in Delhi")
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])
df2['Hour'] = df2['Date'].dt.hour
hourly_pm25dha = df2.groupby('Hour')['PM2.5, ug/m3'].mean()

hourly_pm25dha.plot(kind='line')
plt.ylabel('PM2.5, ug/m3')
plt.title ("Hourly PM 2.5's Level in Dhaka")
```

```python
df3['Date'] = pd.to_datetime(df3['Date'])
df3['Hour'] = df3['Date'].dt.hour
hourly_pm25la = df3.groupby('Hour')['PM2.5, ug/m3'].mean()

hourly_pm25la.plot(kind='line')
plt.ylabel('PM2.5, ug/m3')
plt.title ("Hourly PM 2.5's Level in Lahore")
```
